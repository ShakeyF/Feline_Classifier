```python
from PIL import Image, ImageFilter
import random
```

```python
image = Image.open('raw/0000.jpg')
width,height = image.size
image
```

```python
percentage = 0.7 #percentage of blured part
w = random.randint(0, width-int(percentage*width))
h = random.randint(0, height-int(percentage*height))
```

```python
box = (w,h,w+int(percentage*width),h+int(percentage*height))
```

```python
ic = image.crop(box)
for i in range(100): 
    ic = ic.filter(ImageFilter.BLUR)
image.paste(ic, box)
```
